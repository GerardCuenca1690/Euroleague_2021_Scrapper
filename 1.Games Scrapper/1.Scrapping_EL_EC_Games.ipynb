```python
#Importing libraries

import pandas as pd
pd.set_option('display.max_columns', None)
import numpy as np

import requests
import warnings
warnings.filterwarnings(action="ignore")
```

<font color='black'><h3><left> 1. Parameters set-up </left></h3></font>

* Firstly, we need to set-up the parameters to run our script: season, number of matches and competition.
* **season**: Requires a single integer value. The number you assign to this variable stands for the initial year of the season (2020 will access data from 2020-21 season).
* **matches**: Number of matches each season has. To find the number of matches of a given season you can access the Final Game link of each one at Euroleague/Eurocup website. e.g: Season 2020-2021 had 328 games: https://www.euroleague.net/main/results/showgame?gamecode=328&seasoncode=E2020
* **competition**: Expecting a string variable -> 'E' for Euroleague and 'U' for Eurocup.

```python
#Setting up the parameters of our download:
season=2020
matches=328
competition='E'
```

<font color='black'><h3><left> 2. The scrapper itself </left></h3></font>

The following cell is ready to scrap all the games you want from a given season according to the criteria you indicated on the previous one.

In a nutshell:
* We **generate the URL** for each game we need through concatening different variables (url_string_concat).
* We **append URLs** in url_list
* Using requests we **normalize each json (json_normalize) in an object call Game_Meta**. Last, but not least, we collect each Game_Meta in a **unique DataFrame called Game_Info**.


```python
#Accessing Euroleague API: json Game Info files

########################################################################
#------------------------SCRAPPING LOOP--------------------------------#
########################################################################
url_list = []



number_list = list(range(1,matches+1)) 
for i in number_list:
        url_string_concat = 'https://live.euroleague.net/api/Header?gamecode=' + str(i) + '&seasoncode=' + competition + str(season) + '&disp='
        url_list.append(url_string_concat)

        
#Appending each Match info file into a single Dataframe
Game_Info=pd.DataFrame()
for x in url_list:
        try:
            r = requests.get(x).json()
            Game_Meta = pd.json_normalize(r)
            Game_Meta['LINK']=x
            Game_Info=Game_Info.append(Game_Meta, ignore_index = True)
            print(x)
            del Game_Meta
            del x 
        except:
            pass
        
########################################################################
#Fine-Tuning: Creating new variables + Reordering our Game_Info dataset#
########################################################################  


#Creating a unique GAME_ID for each match. This will allow us merging the Game_Info dataframe
#with other info from the page such as PlayByPlay data or the Shooting Chart
Game_Info['A'], Game_Info['B'] = Game_Info['LINK'].str.split('&', 1).str
Game_Info['A'], Game_Info['C'] = Game_Info['A'].str.split('=', 1).str
Game_Info['D'] = Game_Info['B'].str[11:16]
Game_Info['GAME_ID']=Game_Info['D']+'_'+Game_Info['C'] 
Game_Info['Game']=Game_Info['C']


#Selecting the columns we want to preserve in the final Dataframe
Game_Info=Game_Info[['GAME_ID','Round','Game','Date','Hour','Stadium','Capacity','TeamA','TeamB',
                   'ScoreA','ScoreB','CoachA','CoachB','GameTime','FoultsA','FoultsB',
                   'TimeoutsA','TimeoutsB','ScoreQuarter1A','ScoreQuarter2A','ScoreQuarter3A','ScoreQuarter4A','ScoreExtraTimeA',
                    'ScoreQuarter1B','ScoreQuarter2B','ScoreQuarter3B','ScoreQuarter4B','ScoreExtraTimeB', 
                     'Phase','Competition','CompetitionReducedName','Referee1','Referee2','Referee3']]


###NEW FEATURES

#Match
Game_Info['Match']=Game_Info['TeamA']+' vs '+Game_Info['TeamB']




#To Calculate Results we first need to reconvert ScoreA and ScoreB into integers
Game_Info['ScoreA']=Game_Info['ScoreA'].astype(int)
Game_Info['ScoreB']=Game_Info['ScoreB'].astype(int)
#ResultA
Game_Info['ResultA']=Game_Info['ScoreA']-Game_Info['ScoreB']
#ResultB
Game_Info['ResultB']=Game_Info['ScoreB']-Game_Info['ScoreA']



Game_Info.head()
```

<font color='black'><h3><left> 3. Reshaping Game_Info to create Games, our final dataframe: </left></h3></font>

We don't like how the dataframe looks like in terms of structure directly from EL/EC API. Each row contains info from both contenders and we believe reshaping it generating a single row per team will improve the analysis stage.

* To do this: We generate two new auxiliar df (Games_A & Games_B) containing data from Home and Away teams respectively and then we concatenate them again in a single df (actually the final one) called **Games**.

* Besides, we create more features that will help us filtering as well as generating insights: **They are: Location, W/L or PointsQuarter1, PointsQuarter2, PointsQuarter3, PointsQuarter4, PointsExtraTime**


```python
#1 dataset per Home Team, 1 dataset per Away Team + Final Concat


###HOME TEAM

#Selecting Columns
Games_A=Game_Info[['GAME_ID','Round','Game',
                   'Date','Hour','Stadium','Capacity','Match','TeamA',
                   'ScoreA','ResultA','CoachA','GameTime','FoultsA',
                   'TimeoutsA',
                   'ScoreQuarter1A','ScoreQuarter2A','ScoreQuarter3A','ScoreQuarter4A','ScoreExtraTimeA',
                    'Phase','Competition','CompetitionReducedName','Referee1','Referee2','Referee3']]

#Adding Location Column
Games_A['Location']='Home'

#Renaming A specific columns to allow the future concat.
Games_A= Games_A.rename(columns=
                      {'TeamA': 'Team', 
                       'ScoreA': 'Score',
                       'ResultA': 'Result',
                       'CoachA': 'Coach',
                       'FoultsA': 'Foults',
                       'TimeoutsA': 'Timeouts',
                       'ScoreQuarter1A': 'ScoreQuarter1',
                       'ScoreQuarter2A': 'ScoreQuarter2',
                       'ScoreQuarter3A': 'ScoreQuarter3',
                       'ScoreQuarter4A': 'ScoreQuarter4',
                       'ScoreExtraTimeA': 'ScoreExtraTime'})


###AWAY TEAM

#Selecting Columns
Games_B=Game_Info[['GAME_ID','Round','Game',
                   'Date','Hour','Stadium','Capacity','Match','TeamB',
                   'ScoreB','ResultB','CoachB','GameTime','FoultsB',
                   'TimeoutsB',
                   'ScoreQuarter1B','ScoreQuarter2B','ScoreQuarter3B','ScoreQuarter4B','ScoreExtraTimeB',
                    'Phase','Competition','CompetitionReducedName','Referee1','Referee2','Referee3']]

#Adding Location Column
Games_B['Location']='Away'

#Renaming B specific columns to allow the future concat.
Games_B=Games_B.rename(columns=
                      {'TeamB': 'Team', 
                       'ScoreB': 'Score',
                      'ResultB': 'Result',
                       'CoachB': 'Coach',
                       'FoultsB': 'Foults',
                       'TimeoutsB': 'Timeouts',
                       'ScoreQuarter1B': 'ScoreQuarter1',
                       'ScoreQuarter2B': 'ScoreQuarter2',
                       'ScoreQuarter3B': 'ScoreQuarter3',
                       'ScoreQuarter4B': 'ScoreQuarter4',
                       'ScoreExtraTimeB': 'ScoreExtraTime'})

#Concatening both objects in Games to complete reshaping operation
Games=pd.concat([Games_A,Games_B])

```

```python
###MORE CLEANING OPERATIONS

#Foults/TimeOuts/Round/Capacity type transformation (as integers)
Games['Foults']=Games['Foults'].astype(int, errors='ignore')
Games['Timeouts']=Games['Timeouts'].astype(int, errors='ignore')
Games['Round']=Games['Round'].astype(int, errors='ignore')
Games['Capacity']=Games['Capacity'].astype(int, errors='ignore')
Games['Game']=Games['Game'].astype(int, errors='ignore')


###NEW VARIABLES TO ENRICH OUR FINAL DF

#Win/Loss
Games['W/L']=''
Games['W/L'][Games['Result']>0]='W'
Games['W/L'][Games['Result']<0]='L'

#Points x Quarter
Games['PointsQuarter1']=Games['ScoreQuarter1']
Games['PointsQuarter2']=Games['ScoreQuarter2']-Games['ScoreQuarter1']
Games['PointsQuarter3']=Games['ScoreQuarter3']-Games['ScoreQuarter2']
Games['PointsQuarter4']=Games['ScoreQuarter4']-Games['ScoreQuarter3']

Games['PointsExtra']=Games['ScoreExtraTime']-Games['ScoreQuarter4']
Games['PointsExtraTime']=''
Games['PointsExtraTime'][Games['PointsExtra']>0]=Games['PointsExtra']
Games['PointsExtraTime'][Games['PointsExtra']<=0]=0



###FINAL REORDER + SORTING OUR DF x ROUND&GAME
Games=Games[['GAME_ID','Round','Game',
             'Date','Hour','Stadium','Capacity','Match','Team', 'Location',
                   'Score','Result','W/L','Coach','GameTime','Foults',
                   'Timeouts',
                   'ScoreQuarter1','ScoreQuarter2','ScoreQuarter3','ScoreQuarter4','ScoreExtraTime',
                    'PointsQuarter1','PointsQuarter2','PointsQuarter3','PointsQuarter4','PointsExtraTime',
                    'Phase','Competition','CompetitionReducedName','Referee1','Referee2','Referee3']]

#Sort final df
Games=Games.sort_values(by=['Round','Game'])
Games.head()
```

<font color='black'><h3><left> 4. Saving Games into an Excel file </left></h3></font>

We generate an Excel file containing the Games df. It will help us analyzing it in Excel or more sophisticated BI tools such as Power BI or Tableau.

```python
#Firstly, we create comp variable based on competition to generate a clear output file naming
if competition=='E':
    comp='Euroleague'
else:
    comp='Eurocup'

#Saving it to Excel
Games.to_excel('Games'+'_'+comp+'_'+str(season)+'.xlsx', index=False, sheet_name='Game_Info')
```
