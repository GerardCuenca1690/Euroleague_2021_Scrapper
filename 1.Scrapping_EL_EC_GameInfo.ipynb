```python
#Importando librer√≠as

import pandas as pd
pd.set_option('display.max_columns', None)
import numpy as np

import requests
import warnings
warnings.filterwarnings(action="ignore")
```

```python
#Introducing parameters to download data

#Season
season=2021
matches=21
competition='U'
```

```python
#Accessing Euroleague API: json Match Information files


url_list = []



number_list = list(range(1,matches)) 
for i in number_list:
        url_string_concat = 'https://live.euroleague.net/api/Header?gamecode=' + str(i) + '&seasoncode=' + competition + str(season) + '&disp='
        url_list.append(url_string_concat)

        
#Appending each Match info file into a single Dataframe
Game_Info=pd.DataFrame()
for x in url_list:
        r = requests.get(x).json()
        Game_Meta = pd.json_normalize(r)
        Game_Meta['LINK']=x
        Game_Info=Game_Info.append(Game_Meta, ignore_index = True)
        print(x)
        del Game_Meta
        del x 

#Creating auxiliar variables in order to merge with the other Dataf rames later on        
Game_Info['A'], Game_Info['B'] = Game_Info['LINK'].str.split('&', 1).str
Game_Info['A'], Game_Info['C'] = Game_Info['A'].str.split('=', 1).str
Game_Info['D'] = Game_Info['B'].str[11:16]
Game_Info['GAME_ID']=Game_Info['D']+'_'+Game_Info['C'] 
Game_Info['Game']=Game_Info['C']


#Selecting the columns we want to preserve in the final Dataframe
Game_Info=Game_Info[['GAME_ID','Round','Game','Date','Hour','Stadium','Capacity','TeamA','TeamB',
                   'ScoreA','ScoreB','CoachA','CoachB','GameTime','FoultsA','FoultsB',
                   'TimeoutsA','TimeoutsB',
                'ScoreQuarter1A','ScoreQuarter2A','ScoreQuarter3A','ScoreQuarter4A','ScoreExtraTimeA',
                    'ScoreQuarter1B','ScoreQuarter2B','ScoreQuarter3B','ScoreQuarter4B','ScoreExtraTimeB', 
                    'Phase','Competition','CompetitionReducedName','Referee1','Referee2','Referee3']]
#Match
Game_Info['Match']=Game_Info['TeamA']+' vs '+Game_Info['TeamB']


Game_Info['ScoreA']=Game_Info['ScoreA'].astype(int)
Game_Info['ScoreB']=Game_Info['ScoreB'].astype(int)

#ResultA
Game_Info['ResultA']=Game_Info['ScoreA']-Game_Info['ScoreB']
#ResultB
Game_Info['ResultB']=Game_Info['ScoreB']-Game_Info['ScoreA']


Game_Info
```

```python
#1 dataset per Home Team, 1 dataset per Away Team + concat


###HOME
Games_A=Game_Info[['GAME_ID','Round','Game',
                   'Date','Hour','Stadium','Capacity','Match','TeamA',
                   'ScoreA','ResultA','CoachA','GameTime','FoultsA',
                   'TimeoutsA',
                   'ScoreQuarter1A','ScoreQuarter2A','ScoreQuarter3A','ScoreQuarter4A','ScoreExtraTimeA',
                    'Phase','Competition','CompetitionReducedName','Referee1','Referee2','Referee3']]

Games_A['Location']='Home'
Games_A= Games_A.rename(columns=
                      {'TeamA': 'Team', 
                       'ScoreA': 'Score',
                       'ResultA': 'Result',
                       'CoachA': 'Coach',
                       'FoultsA': 'Foults',
                       'TimeoutsA': 'Timeouts',
                       'ScoreQuarter1A': 'ScoreQuarter1',
                       'ScoreQuarter2A': 'ScoreQuarter2',
                       'ScoreQuarter3A': 'ScoreQuarter3',
                       'ScoreQuarter4A': 'ScoreQuarter4',
                       'ScoreExtraTimeA': 'ScoreExtraTime'})


###AWAY
Games_B=Game_Info[['GAME_ID','Round','Game',
                   'Date','Hour','Stadium','Capacity','Match','TeamB',
                   'ScoreB','ResultB','CoachB','GameTime','FoultsB',
                   'TimeoutsB',
                   'ScoreQuarter1B','ScoreQuarter2B','ScoreQuarter3B','ScoreQuarter4B','ScoreExtraTimeB',
                    'Phase','Competition','CompetitionReducedName','Referee1','Referee2','Referee3']]

Games_B['Location']='Away'
Games_B=Games_B.rename(columns=
                      {'TeamB': 'Team', 
                       'ScoreB': 'Score',
                      'ResultB': 'Result',
                       'CoachB': 'Coach',
                       'FoultsB': 'Foults',
                       'TimeoutsB': 'Timeouts',
                       'ScoreQuarter1B': 'ScoreQuarter1',
                       'ScoreQuarter2B': 'ScoreQuarter2',
                       'ScoreQuarter3B': 'ScoreQuarter3',
                       'ScoreQuarter4B': 'ScoreQuarter4',
                       'ScoreExtraTimeB': 'ScoreExtraTime'})

#Concat
Games=pd.concat([Games_A,Games_B])

#Foults/TimeOuts/Round/Capacity Transformation
Games['Foults']=Games['Foults'].astype(int)
Games['Timeouts']=Games['Timeouts'].astype(int)
Games['Round']=Games['Round'].astype(int)
Games['Capacity']=Games['Capacity'].astype(int)
Games['Game']=Games['Game'].astype(int)

#Win/Loss
Games['W/L']=''
Games['W/L'][Games['Result']>0]='W'
Games['W/L'][Games['Result']<0]='L'

#Points x Quarter
Games['PointsQuarter1']=Games['ScoreQuarter1']
Games['PointsQuarter2']=Games['ScoreQuarter2']-Games['ScoreQuarter1']
Games['PointsQuarter3']=Games['ScoreQuarter3']-Games['ScoreQuarter2']
Games['PointsQuarter4']=Games['ScoreQuarter4']-Games['ScoreQuarter3']

Games['PointsExtra']=Games['ScoreExtraTime']-Games['ScoreQuarter4']
Games['PointsExtraTime']=''
Games['PointsExtraTime'][Games['PointsExtra']>0]=Games['PointsExtra']
Games['PointsExtraTime'][Games['PointsExtra']<=0]=0

Games=Games[['GAME_ID','Round','Game',
             'Date','Hour','Stadium','Capacity','Match','Team', 'Location',
                   'Score','Result','W/L','Coach','GameTime','Foults',
                   'Timeouts',
                   'ScoreQuarter1','ScoreQuarter2','ScoreQuarter3','ScoreQuarter4','ScoreExtraTime',
                    'PointsQuarter1','PointsQuarter2','PointsQuarter3','PointsQuarter4','PointsExtraTime',
                    'Phase','Competition','CompetitionReducedName','Referee1','Referee2','Referee3']]

#Sort final df
Games=Games.sort_values(by=['Round','Game'])
Games
```
